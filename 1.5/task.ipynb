
### Варіант 1: Облік книг у бібліотеці
#### Завдання:
1. **Проектування бази даних**:
   - Таблиця `books` з полями:
     - `book_id`: унікальний ідентифікатор книги (SERIAL).
     - `title`: назва книги (VARCHAR).
     - `author`: автор книги (VARCHAR).

2. **Розробка REST API**:
   - `GET /api/books`: отримати всі книги.
   - `POST /api/books`: додати нову книгу.
   - `DELETE /api/books/<book_id>`: видалити книгу за `book_id`.

3. **Розробка простого UI**:
   - Веб-сторінка для відображення книг.
   - Форма для додавання нової книги.
   - Функціонал для видалення книги.



### Загальні вимоги для всіх варіантів:
- Використання PostgreSQL для бази даних.
- Розробка REST API за допомогою Flask.
- Створення простого користувацького інтерфейсу з HTML, CSS і JavaScript.
- Створити окрему папку для проекту в існуючому репозиторії GitHub, з назвою practice5.

### Очікувані результати:
Слухачі повинні вміти створювати просту ІС з базою даних, розробляти REST API та реалізувати базовий веб-інтерфейс для взаємодії з даними, обравши один з варіантів.
