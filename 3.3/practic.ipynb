```python
import pandas as pd
```

```python
df = pd.read_csv('global firepower 2022 wide.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.describe())
```

```python

import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Tanks'], bins=15, color='skyblue', edgecolor='black')
plt.title('Розподіл кількості танків')
plt.xlabel('Кількість танків')
plt.ylabel('Кількість країн')
plt.grid(True)
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(
    [df['Tanks'].dropna(), df['Total Aircraft Strength'].dropna(), df['Active Personnel'].dropna()],
    labels=['Tanks', 'Aircraft', 'Personnel'],
    patch_artist=True,
    boxprops=dict(facecolor='lightgreen', color='darkgreen'),
    medianprops=dict(color='red', linewidth=2)
)
plt.title('Порівняння розподілу військових показників')
plt.ylabel('Кількість одиниць')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df['Defense Budget'], df['Active Personnel'], color='blue', alpha=0.6, edgecolors='black')
plt.title('Залежність між оборонним бюджетом і чисельністю персоналу')
plt.xlabel('Defense Budget')
plt.ylabel('Active Personnel')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Коефіцієнт кореляції')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Кореляційна матриця показників')
plt.tight_layout()
plt.show()
```
