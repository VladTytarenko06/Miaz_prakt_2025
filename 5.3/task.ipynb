# **10 варіантів практичних завдань**

## **Варіант 1. Побудова карти тактичної обстановки з нанесенням подій**

**Завдання:**
На основі CSV із подіями (час, координати, тип події, інтенсивність) побудувати інтерактивну карту у Python (Folium) або Power BI (якщо доступно).
**Очікуваний результат:**
Карта з маркерами, градацією кольору за типами подій, можливістю фільтрування за часом.

---

## **Варіант 2. Побудова часових рядів динаміки ворожих атак**

**Завдання:**
За таблицею (добові кількості атак за напрямками) побудувати 2 графіки:

1. лінійний графік;
2. стовпчикову порівняльну діаграму по напрямках.
   **Очікуваний результат:**
   Виявлені піки активності та спад, зроблено короткі аналітичні висновки.

---

## **Варіант 3. Візуалізація структури сил противника**

**Завдання:**
Побудувати кругову (або Treemap) діаграму для структури сил противника: піхота, бронетехніка, артилерія, БПЛА тощо.
**Очікуваний результат:**
Коректно відображена структура у %, коротке обґрунтування, яке угруповання переважає.

---

## **Варіант 4. Порівняння боєздатності підрозділів (оцінка readiness)**

**Завдання:**
Побудувати радар-чарт (Radar/Spider chart) для 3–5 підрозділів за параметрами:

* особовий склад
* техніка
* забезпечення
* управління
* морально-психологічний стан
  **Очікуваний результат:**
  Графік + аналітичний висновок щодо відхилень.

---

## **Варіант 5. Візуалізація мережевої взаємодії (Network Graph)**

**Завдання:**
Побудувати мережевий граф взаємодії між підрозділами (J-структура, підпорядкування, канали передачі інформації).
**Очікуваний результат:**
NetworkGraph, визначення центральних вузлів, слабких місць інформаційних потоків.

---

## **Варіант 6. Кореляційний аналіз факторів, що впливають на втрати противника**

**Завдання:**
Побудувати heatmap кореляцій: інтенсивність атак, погода, відстань до переднього краю, час доби, тип озброєння.
**Очікуваний результат:**
Теплова карта + коротка інтерпретація.

---

## **Варіант 7. Побудова дашборду оперативного огляду обстановки**

**Завдання:**
Інтегрувати 3–5 візуалізацій (карта, часовий ряд, структура сил, таблиця подій) у єдиний дашборд (Power BI / Tableau / Python Dash).
**Очікуваний результат:**
Прототип дашборда з мінімальною взаємодією.

---

## **Варіант 8. Порівняння результативності БПЛА**

**Завдання:**
Побудувати boxplot або violin plot за показниками ефективності різних типів БПЛА (дальність, точність, час польоту, вартість).
**Очікуваний результат:**
Графік + висновки щодо варіативності та середніх значень.

---

## **Варіант 9. Аналіз маршруту логістичного забезпечення**

**Завдання:**
Побудувати лінійний графік + картографічну візуалізацію маршруту з перешкодами, ризиками, завантаженням.
**Очікуваний результат:**
Маршрут, ключові точки ризику, рекомендації.

---

## **Варіант 10. Візуалізація сценаріїв прогнозування**

**Завдання:**
Побудувати 3 лінії прогнозу на основі синтетичних даних: оптимістичний, базовий, песимістичний сценарій.
**Очікуваний результат:**
Графік + коротка інтерпретація, виділення зони невизначеності.

# **Детальна методика виконання завдання (універсальна для всіх 10 варіантів)**

Методика подається у форматі покрокового військово-аналітичного підходу, який можна застосувати до кожного варіанту.

---

# **1. Підготовчий етап**

### **1.1. Аналіз вихідних даних**

Слухач повинен:

* визначити формат даних (CSV, Excel, JSON, координати тощо);
* перевірити повноту, пропуски, дублікати;
* оцінити які змінні доречні для візуалізації.

### **1.2. Вибір доцільного типу візуалізації**

Вибір залежить від:

* характеру даних (категоріальні, кількісні, гео-дані);
* задачі аналізу (порівняння, структура, динаміка, кореляції, прогноз);
* рівня командування (оперативний — наголос на швидкому сприйнятті).

---

# **2. Технічний етап**

### **2.1. Завантаження та підготовка даних**

* імпорт даних у Python (Pandas) або Power BI;
* очищення: заповнення пропусків, фільтрування, нормалізація;
* якщо є геокоординати — перевести у формат latitude/longitude.

### **2.2. Побудова базової візуалізації**

Для Python (matplotlib/plotly/Folium):

* створити об’єкт фігури;
* побудувати основний графік;
* налаштувати колір, масштаб, підписи, легенду, grid.

Для Power BI:

* використати відповідний візуальний елемент;
* додати поля;
* налаштувати фільтри та слайсери.

---

# **3. Аналітичний етап**

### **3.1. Виявлення ключових патернів**

* тренди (зростання/спад);
* піки активності;
* найбільш проблемні напрямки;
* «аномалії» (викиди, різкі стрибки).

### **3.2. Формування аналітичних висновків**

Слухач має сформувати 3–5 чітких висновків:

* що показує графік;
* які елементи привертають увагу;
* які фактори можуть пояснювати дані;
* які дії або рішення можуть випливати для штабу.

---

# **4. Оформлення результатів**

### **4.1. Структура звіту**

1. Короткий опис завдання
2. Характеристика вихідних даних
3. Обраний тип візуалізації (чому саме він)
4. Побудований графік / карта / дашборд
5. Аналітичні висновки
6. Рекомендації (за необхідності)

### **4.2. Вимоги до якості**

* кольорова палітра військового стандарту (olive/earth tones/NATO blue);
* мінімум зайвої інформації на графіку;
* читабельні підписи;
* зрозумілий масштаб;
* короткі пояснення.

---

# **5. Контрольні запитання для самоперевірки**
