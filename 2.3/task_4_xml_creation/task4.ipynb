```python
import xml.etree.ElementTree as ET

def create_unit_xml(name, quantity, status, filename="unit.xml"):
    # Створюємо кореневий елемент
    unit = ET.Element("Unit")

    # Додаємо під-елементи
    name_elem = ET.SubElement(unit, "Name")
    name_elem.text = name

    quantity_elem = ET.SubElement(unit, "Quantity")
    quantity_elem.text = str(quantity)

    status_elem = ET.SubElement(unit, "Status")
    status_elem.text = status

    # Створюємо дерево
    tree = ET.ElementTree(unit)

    # Записуємо у файл
    tree.write(filename, encoding="utf-8", xml_declaration=True)
    print(f"XML-файл '{filename}' успішно створено.")

if __name__ == "__main__":
    # Приклад даних
    create_unit_xml("1-й батальйон", 320, "боєздатний", "unit.xml")
```
